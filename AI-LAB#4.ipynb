```python
#Task 1
graph = {
    'Tehran': ['Baghdad', 'Istanbul'],
    'Baghdad': ['Cairo'],
    'Istanbul': ['Berlin'],
    'Cairo': ['Washington'],
    'Berlin': ['Washington']
}

def bfs_peace_drone(graph, start, goal):
    visited = [] 
    queue = [[start]] # Store the full path in the queue

    while queue:
        path = queue.pop(0) # FIFO
        node = path[-1]

        if node not in visited:
            print(f"Visiting: {node}")
            if node == goal:
                return path
            visited.append(node)
            
            for neighbor in graph.get(node, []):
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)

    return "Peace treaty lost."

# Execution
start_node = 'Tehran'
goal_node = 'Washington'
final_path = bfs_peace_drone(graph, start_node, goal_node)
print("\nWorld Saved! Path taken:", " -> ".join(final_path))
```

```python
#Task 2
graph = {
    'Earth': {'Moon_Base': 10, 'Orbital_Platform': 5},
    'Orbital_Platform': {'Moon_Base': 2, 'Mars': 60},
    'Moon_Base': {'Mars': 50},
    'Mars': {}
}

def ucs_mars(graph, start, goal):
    # (cost, path)
    frontier = [(0, [start])]
    visited = set()

    while frontier:
        # Simulate priority queue by sorting by cost [cite: 769]
        frontier.sort(key=lambda x: x[0])
        cost, path = frontier.pop(0)
        current = path[-1]

        if current in visited:
            continue
        
        if current == goal:
            return path, cost
        
        visited.add(current)

        for neighbor, fuel in graph[current].items():
            if neighbor not in visited:
                new_path = list(path)
                new_path.append(neighbor)
                frontier.append((cost + fuel, new_path))

    return None, float('inf')

# Execution
path, total_fuel = ucs_mars(graph, 'Earth', 'Mars')
print(f"Cheapest path to Mars: {' -> '.join(path)}")
print(f"Total fuel cost: {total_fuel} units")
```

```python
#Task 3
graph = {
    'Entrance': ['Hallway_A', 'Hallway_B'],
    'Hallway_A': ['Storage'],
    'Hallway_B': ['Target'],
    'Storage': ['Deep_Vault'],
    'Target': [],
    'Deep_Vault': []
}

def dls_bunker(graph, current, goal, limit, depth, visited):
    print(f"Searching {current} at depth {depth}")
    
    if current == goal:
        return True
    
    if depth >= limit:
        return False # Oxygen limit reached [cite: 654]

    for neighbor in graph.get(current, []):
        if neighbor not in visited:
            visited.append(neighbor)
            if dls_bunker(graph, neighbor, goal, limit, depth + 1, visited):
                return True
            visited.pop() # Backtrack [cite: 664]
            
    return False

# Execution
start_node = 'Entrance'
target_node = 'Target'
limit = 2

if dls_bunker(graph, start_node, target_node, limit, 0, [start_node]):
    print("\nMission Accomplished!")
else:
    print("\nTarget not found within oxygen limit.")
```

```python
#Task 4
grid = {
    'Main_Box': ['Zone_A', 'Zone_B'],
    'Zone_A': ['Switch_1', 'Switch_2'],
    'Zone_B': ['Breaker_101']
}

def dls_check(node, goal, depth):
    if depth == 0 and node == goal:
        return True
    if depth > 0:
        for neighbor in grid.get(node, []):
            if dls_check(neighbor, goal, depth - 1):
                return True
    return False

def ids_restore_power(start, goal, max_d):
    for depth in range(max_d + 1):
        print(f"Checking Depth {depth}...")
        if dls_check(start, goal, depth):
            print(f"Found {goal}! Lights are back on!")
            return True
    return False

# Execution
ids_restore_power('Main_Box', 'Breaker_101', 2)
```
