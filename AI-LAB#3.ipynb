```python
# TASK 1
class Environment:
    def __init__(self):
        self.beds = ["Moist", "Moist", "Dry", "Moist", "Moist", "Dry", "Moist", "Dry", "Moist"]

    def get_percept(self, position):
        return self.beds[position]

    def water_bed(self, position):
        self.beds[position] = "Moist"

class SimpleReflexAgent:
    def act(self, percept):
        if percept == "Dry":
            return "Watering"
        else:
            return "Moving"

def run_simulation():
    e = Environment()
    a = SimpleReflexAgent()
    
    for i in range(9):
        per = e.get_percept(i)
        act= a.act(per)
        
        if act == "Watering":
            e.water_bed(i)
            print(f"Bed {i+1}: Dry, Watered. Now Moist.")
        else:
            print(f"Bed {i+1}: Moist, No action needed.")
            
    print(f"Final Greenhouse Status: {e.beds}")

run_simulation()
```

```python
# TASK 2
class Environment:
    def __init__(self):
        self.path = ['A', 'C', 'D', 'B', 'D']
        self.current_index = 0

    def get_percept(self):
        return self.path[self.current_index]

class ModelBasedAgent:
    def __init__(self):
        self.model = {"has_keycard": False}

    def act(self, room):
        if room == 'B':
            self.model["has_keycard"] = True
            return "Picked up keycard"
        
        if room == 'D':
            if self.model["has_keycard"]:
                return "Access Granted to Vault"
            else:
                return "Access Denied: Keycard required"
        
        return "Patrolling"

def run_patrol():
    env = Environment()
    agent = ModelBasedAgent()
    
    for i in range(5):
        room = env.get_percept()
        action = agent.act(room)
        print(f"Location {room}: {action}")
        env.current_index += 1

run_patrol()
```

```python
# TASK 3
class Environment:
    def __init__(self):
        self.houses = ['Blue House', 'Green House', 'Red House', 'Yellow House', 'White House']

class GoalBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def check_goal(self, current_location):
        if current_location == self.goal:
            return True
        return False

def run_delivery():
    env = Environment()
    agent = GoalBasedAgent("Red House")
    
    for house in env.houses:
        print(f"Drone flying to: {house}")
        if agent.check_goal(house):
            print(f"SUCCESS: Reached the {house}. Mission Terminated.")
            return

run_delivery()
```

```python
# TASK 4
class Environment:
    def __init__(self):
        self.rocks = {
            "Rock A": {"Value": 5, "Cost": 2},
            "Rock B": {"Value": 9, "Cost": 8},
            "Rock C": {"Value": 6, "Cost": 3}
        }

class UtilityBasedAgent:
    def calculate_utility(self, value, cost):
        return (value * 2) - cost

    def select_best_rock(self, rocks):
        best_rock = None
        max_utility = -float('inf')
        
        for name, data in rocks.items():
            u = self.calculate_utility(data["Value"], data["Cost"])
            print(f"{name} Utility: {u}")
            if u > max_utility:
                max_utility = u
                best_rock = name
        return best_rock

def run_rover():
    env = Environment()
    agent = UtilityBasedAgent()
    best = agent.select_best_rock(env.rocks)
    print(f"\nMars Rover selects {best} for sampling.")

run_rover()
```

```python
# TASK 5
import random

class Environment:
    def __init__(self):
        self.spots = ['Start', 'Empty', 'Cheese', 'Trap']
        self.rewards = {'Start': 0, 'Empty': -1, 'Cheese': 10, 'Trap': -10}

    def get_random_move(self):
        return random.choice(self.spots[1:])

class LearningAgent:
    def process_feedback(self, state, reward):
        return f"Mouse moved to {state} and received {reward} points."

def run_learning():
    env = Environment()
    agent = LearningAgent()
    
    target_state = env.get_random_move()
    reward = env.rewards[target_state]
    
    result = agent.process_feedback(target_state, reward)
    print(result)

run_learning()
```
