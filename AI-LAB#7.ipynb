```python
from ortools.sat.python import cp_model

def solve_shift_assignment():
    model = cp_model.CpModel()
    alice   = model.new_int_var(0, 2, "Alice")
    bob     = model.new_int_var(0, 2, "Bob")
    charlie = model.new_int_var(0, 2, "Charlie")

    model.add_all_different([alice, bob, charlie])
    model.add(alice != 0)
    solver = cp_model.CpSolver()
    status = solver.solve(model)
    shift_names = {0: "Morning", 1: "Afternoon", 2: "Evening"}

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print(f"  Alice   → {shift_names[solver.value(alice)]}")
        print(f"  Bob     → {shift_names[solver.value(bob)]}")
        print(f"  Charlie → {shift_names[solver.value(charlie)]}")
    else:
        print("No solution found.")

solve_shift_assignment()
```

```python
from ortools.sat.python import cp_model

def solve_send_more_money():
    model = cp_model.CpModel()
    S = model.new_int_var(0, 9, "S")
    E = model.new_int_var(0, 9, "E")
    N = model.new_int_var(0, 9, "N")
    D = model.new_int_var(0, 9, "D")
    M = model.new_int_var(0, 9, "M")
    O = model.new_int_var(0, 9, "O")
    R = model.new_int_var(0, 9, "R")
    Y = model.new_int_var(0, 9, "Y")

    letters = [S, E, N, D, M, O, R, Y]
    
    model.add_all_different(letters)
    
    model.add(S != 0)
    model.add(M != 0)
    
    model.add(
        1000 * S + 100 * E + 10 * N + D
        + 1000 * M + 100 * O + 10 * R + E
        == 10000 * M + 1000 * O + 100 * N + 10 * E + Y
    )

    solver = cp_model.CpSolver()
    status = solver.solve(model)
    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        s_val = solver.value(S)
        e_val = solver.value(E)
        n_val = solver.value(N)
        d_val = solver.value(D)
        m_val = solver.value(M)
        o_val = solver.value(O)
        r_val = solver.value(R)
        y_val = solver.value(Y)

        send  =  1000*s_val + 100*e_val + 10*n_val + d_val
        more  =  1000*m_val + 100*o_val + 10*r_val + e_val
        money = 10000*m_val + 1000*o_val + 100*n_val + 10*e_val + y_val

        print("  SEND + MORE = MONEY — Solution Found")
        print(f"  S = {s_val}")
        print(f"  E = {e_val}")
        print(f"  N = {n_val}")
        print(f"  D = {d_val}")
        print(f"  M = {m_val}")
        print(f"  O = {o_val}")
        print(f"  R = {r_val}")
        print(f"  Y = {y_val}")
        print(f"  {send} + {more} = {money}  ✓" if send + more == money else "  Verification FAILED")
    else:
        print("No solution found.")

solve_send_more_money()
```

```python
from ortools.sat.python import cp_model

def solve_factory_ilp():
    model = cp_model.CpModel()
    MAX_UNITS = 100

    laptops = model.new_int_var(10, MAX_UNITS, "Laptops")
    phones  = model.new_int_var(10, MAX_UNITS, "Phones")
    tablets = model.new_int_var(10, MAX_UNITS, "Tablets")

    model.add(2 * laptops + 1 * phones + 1 * tablets <= 100)

    model.add(1 * laptops + 1 * phones + 1 * tablets <= 80)

    model.add(1 * laptops + 1 * phones + 2 * tablets <= 120)

    model.maximize(500 * laptops + 300 * phones + 400 * tablets)

    solver = cp_model.CpSolver()
    status = solver.solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        l_val = solver.value(laptops)
        p_val = solver.value(phones)
        t_val = solver.value(tablets)
        profit = 500 * l_val + 300 * p_val + 400 * t_val

        print("  Factory ILP — Optimal Production Plan")
        print(f"  Laptops produced : {l_val} units")
        print(f"  Phones produced  : {p_val} units")
        print(f"  Tablets produced : {t_val} units")
        print(f"  Maximum Daily Profit : ${profit:,}")

        chips_used    = 2*l_val + p_val + t_val
        screens_used  = l_val + p_val + t_val
        batteries_used = l_val + p_val + 2*t_val
        print("\n  Resource Usage:")
        print(f"    Microchips : {chips_used}/100")
        print(f"    Screens    : {screens_used}/80")
        print(f"    Batteries  : {batteries_used}/120")
    else:
        print("No feasible solution found.")

solve_factory_ilp()
```
